```python
data['Year_Birth'] = data['Year_Birth'] + 20
```

```python
data['Age'] = date.today().year - data['Year_Birth']
```

```python
def age_category(data):
    if data['Age'] < 13:
        val = 'Kids'
    elif 13 <= data['Age'] < 20:
        val = 'Teens'
    elif 20 <= data['Age'] < 30:
        val = 'Young Adults'
    elif 30 <= data['Age'] < 60:
        val = 'Adults'
    else:
        val = 'Seniors'
        
    return val
```

```python
data['Age_category'] = data.apply(age_category, axis=1)
```

```python
data['Spending']=data['MntWines']+data['MntFruits']+data['MntMeatProducts']+data['MntFishProducts']+data['MntSweetProducts']+data['MntGoldProds']
```

```python
mini = data['Spending'].min()
quater = data['Spending'].quantile(.25)
quater3 = data['Spending'].quantile(.75)
maxi = data['Spending'].max()
```

```python
def spending(data):
    if mini <= data['Spending'] < quater:
        val = 'Low Budget'
    elif quater <= data['Spending'] < quater3:
        val = 'Medium Budget'
    else:
        val = 'High Budget'
    return val
```

```python
data['Budget'] = data.apply(spending, axis=1)
```
